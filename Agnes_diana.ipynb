```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
data=pd.read_csv('advertising.csv')
data.head()
```

```python
data.drop(['TV','Radio'],axis=1,inplace=True)
data
```

```python
data.dropna(inplace=True)
data
```

```python
dend1 = shc.dendrogram(shc.linkage(data,method='ward', metric='euclidean'))
```

```python
agnes=AgglomerativeClustering(n_clusters=5, linkage='ward', metric='euclidean')
agnes.fit(data)
```

```python
data_labels=agnes.labels_
sns.scatterplot(x='Newspaper', y='Sales', data=data, hue=data_labels).set_title('Agnes')
```

```python
dend2 = shc.dendrogram(shc.linkage(data,method='complete', metric='euclidean'))
```

```python
diana=AgglomerativeClustering(n_clusters=5, linkage='complete', metric='euclidean')
diana.fit(data)
```

```python
data_labels=diana.labels_
sns.scatterplot(x='Newspaper', y='Sales', data=data, hue=data_labels).set_title('diana')
```
